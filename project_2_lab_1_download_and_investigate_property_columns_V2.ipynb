```python
import re
year_pattern = re.compile(r'(\d{4})_metro_tax_parcels')
getyear = lambda path: (m.group(1) if (m := year_pattern.search(path)) else None)
```

```python
[getyear(path) for path in parcel_paths]
```

```python
schemas = {
    path: {col: pl.String for col in pl.scan_csv(
        path,
        has_header=True,
        ignore_errors=True,
        separator='|'
    ).limit(1).collect().columns}
    for path in parcel_paths
}
schemas
```

```python
indicator_tables = [
    pl.scan_csv(
        path,
        has_header=True,
        separator='|',
        schema=schemas[path]
    ).limit(1)
    .unpivot()
    .rename({'variable': 'column'})
    .drop('value')
    .with_columns(pl.lit(1).alias(getyear(path)))
    .collect()
    for path in parcel_paths
]
indicator_tables
```

```python
join_next = lambda left, right: left.join(right, on='column', how='full').drop('column_right')
```

```python
join_next(indicator_tables[0], indicator_tables[1])
```

```python
from functools import reduce
indicator_table = reduce(join_next, indicator_tables)
indicator_table
```

```python
indicator_table = (
    reduce(join_next, indicator_tables)
    .filter(pl.col('column').is_not_null())
    .fill_null(0)
)
indicator_table
```

```python
indicator_table.select(cs.integer()).sum()
```

```python
from operator import mul
indicator_table = (
    reduce(join_next, indicator_tables)
    .drop('2002')
    .drop('2003')
    .with_columns(
        pl.reduce(mul, cs.integer()).alias('in_all')
    )
    .filter(pl.col('in_all') == 1)
)
indicator_table
```

```python
indicator_table \
    .with_columns(
        pl.reduce(mul, cs.integer()).alias('in_all')
    ) \
    .filter(pl.col('in_all') == 1) \
    .get_column('column') \
    .to_list()
```
